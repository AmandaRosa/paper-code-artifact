```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file and the specific sheet
file_path = './Resultados.xlsx'  # Update with the correct path to your file
sheet_name = 'Resultados_T_40Hz'
speed = '13 Hz'
domain = 'Time'

# Read the Excel file
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Drop rows with NaN values in 'Feature', 'Acc', 'Método', or 'Channel' columns
df_cleaned = df.dropna(subset=['Feature', 'Acc', 'Método', 'Channel'])

# Convert 'Feature' to string, 'Acc' to float, 'Método' and 'Channel' to string if necessary
df_cleaned['Feature'] = df_cleaned['Feature'].astype(str)
df_cleaned['Acc'] = df_cleaned['Acc'].astype(float)
df_cleaned['Método'] = df_cleaned['Método'].astype(str)
df_cleaned['Channel'] = df_cleaned['Channel'].astype(str)

import seaborn as sns

# Create a heatmap for each channel
channels = df_cleaned['Channel'].unique()

fig, axes = plt.subplots(nrows=len(channels), figsize=(15, 10))

for i, channel in enumerate(channels):
    subset = df_cleaned[df_cleaned['Channel'] == channel]
    heatmap_data = subset.pivot_table(index='Feature', columns='Método', values='Acc')
    sns.heatmap(heatmap_data, ax=axes[i], annot=True, cmap='viridis', vmin=0.9, vmax=1)
    axes[i].set_title(f'Heatmap for {channel} | Speed: {speed} | Domain: {domain}')
    axes[i].set_xlabel('Método')
    axes[i].set_ylabel('Feature')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Example data (replace with actual data)
features = ['feature1', 'feature2', 'feature3']
z_scores = np.random.rand(len(features))
correlations = np.random.rand(len(features))

# Simulated model performance data
models = ['MLP', 'KNN', 'SVM']
accuracies = {
    'MLP': np.random.rand(len(features)),
    'KNN': np.random.rand(len(features)),
    'SVM': np.random.rand(len(features))
}
additional_metrics = {
    'MLP': np.random.rand(len(features)) * 100,
    'KNN': np.random.rand(len(features)) * 100,
    'SVM': np.random.rand(len(features)) * 100
}

# Prepare data for plotting
data = []
for model in models:
    for i, feature in enumerate(features):
        data.append({
            'Feature': feature,
            'Model': model,
            'Z-Score': z_scores[i],
            'Accuracy': accuracies[model][i],
            'Correlation': correlations[i],
            'Additional Metric': additional_metrics[model][i]
        })

data_df = pd.DataFrame(data)

# Create the bubble plot
plt.figure(figsize=(12, 8))
sns.scatterplot(
    x='Z-Score', y='Accuracy',
    hue='Model', style='Model', size='Additional Metric',
    sizes=(20, 200), palette='coolwarm', data=data_df, alpha=0.6, edgecolor='w', linewidth=0.5
)

# Add color bar for correlation
norm = plt.Normalize(data_df['Correlation'].min(), data_df['Correlation'].max())
sm = plt.cm.ScalarMappable(cmap='coolwarm', norm=norm)
sm.set_array([])
plt.colorbar(sm, label='Correlation Coefficient')

plt.xlabel('Z-Score')
plt.ylabel('Accuracy')
plt.title('Feature Z-Score vs Accuracy with Correlation Heatmap and Bubble Diameter as F1 Score')

# Optionally, add labels for each point
for i, row in data_df.iterrows():
    plt.text(row['Z-Score'], row['Accuracy'], row['Feature'], fontsize=9, ha='right')

plt.legend(title='Model')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

speed = '40'
dominio = 'Time-Frequency'
file_path = './Resultados.xlsx'
sheet_name = 'Resultados_TF_40Hz'
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Define your methods
methods = ['MLP', 'SVM', 'NaiveBayes', 'DecisionTree', 'RandomForest', 'KNN']

# Get the Set2 colormap
cmap = plt.get_cmap('Set2')

# Number of colors you need
num_colors = len(methods)

# Get the colors from the colormap
set2_colors = [cmap(i) for i in range(num_colors)]

# Map methods to colors
method_to_color = {method: color for method, color in zip(methods, set2_colors)}

# Example to show the assigned colors
for method, color in method_to_color.items():
    print(f"{method}: {color}")

df_cleaned = df.dropna(subset=['Feature', 'Acc', 'Método', 'Channel'])
df_cleaned['Feature'] = df_cleaned['Feature'].astype(str)
df_cleaned['Acc'] = df_cleaned['Acc'].astype(float)
df_cleaned['Método'] = df_cleaned['Método'].astype(str)
df_cleaned['Channel'] = df_cleaned['Channel'].astype(str)

channels = df_cleaned['Channel'].unique()

for channel in channels:
    channel_df = df_cleaned[df_cleaned['Channel'] == channel]
    pivot_df = channel_df.pivot(index='Feature', columns='Método', values='Acc')

    # Create the bar plot with specific colors for each method
    ax = pivot_df.plot(kind='bar', figsize=(12, 7), color=[method_to_color[method] for method in pivot_df.columns])
    plt.xlabel('Feature')
    plt.ylabel('Acc')
    plt.title(f'Feature vs Acc grouped by Método for {channel} in {dominio} Domain')
    plt.xticks(rotation=10)
    plt.grid(axis='y')
    plt.legend(title='Architecture')

    file_name = f'./imagens_artigo/Barras_{dominio}_canal_{channel}_speed_{speed}Hz.png'
    plt.savefig(file_name, dpi=600)
    plt.show()
```

### PARA CADA VELOCIDADE QUAL CANAL OBTEVE MAIOR MÉDIA ACURÁCIA CONSIDERANDO TODAS FEATURES*?

##### *somente as features que obtiveram acima de 90% de acc

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# File path to the Excel file
file_path = './Resultados_graphs.xlsx'

# Load all sheets into a single DataFrame
sheet_names = ['Resultados_TF_40Hz', 'Resultados_TF_13Hz', 'Resultados_F_40Hz', 'Resultados_F_13Hz', 'Resultados_T_40Hz', 'Resultados_T_13Hz']
dataframes = []

for sheet in sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet)
    # Extract domain and speed from the sheet name
    if 'TF' in sheet:
        domain = 'TF'
    elif 'F' in sheet:
        domain = 'F'
    else:
        domain = 'T'
    speed = sheet.split('_')[-1]
    # Add domain and speed columns to the dataframe
    df['Domain'] = domain
    df['Speed'] = speed
    dataframes.append(df)

# Concatenate all dataframes
df_all = pd.concat(dataframes, ignore_index=True)

# Ensuring column names exist and type conversion
df_all['Feature'] = df_all['Feature'].astype(str)
df_all['Acc'] = df_all['Acc'].astype(float)
df_all['Método'] = df_all['Método'].astype(str)
df_all['Channel'] = df_all['Channel'].astype(str)
df_all['Speed'] = df_all['Speed'].astype(str)

# Grouping and calculating mean accuracy by Channel and Speed
mean_acc_by_speed = df_all.groupby(['Channel', 'Speed'])['Acc'].mean().reset_index()

# Define a custom color palette
custom_palette = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2']

# Plotting the mean accuracy by Channel and Speed
fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(x='Speed', y='Acc', hue='Channel', data=mean_acc_by_speed, ax=ax, palette=custom_palette)

# Define font sizes
title_fontsize = 20
label_fontsize = 16
tick_label_fontsize = 14
legend_fontsize = 14

# Set the title and labels
plt.title('Mean Accuracy by Channel and Speed', fontsize=title_fontsize)
plt.xlabel('Speed', fontsize=label_fontsize, weight= 'bold')
plt.ylabel('Mean Accuracy', fontsize=label_fontsize, weight='bold')

# Adjust legend
plt.legend(title='Channel', fontsize=legend_fontsize, title_fontsize=legend_fontsize, bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust tick label sizes
ax.tick_params(axis='both', which='major', labelsize=tick_label_fontsize)

plt.tight_layout()
plt.show()

file_name = './imagens_artigo/Mean_Accuracy_by_Channel_and_speed.png'
fig.savefig(file_name, dpi=600)
plt.close(fig)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches

# File path to the Excel file
file_path = './Resultados_graphs.xlsx'

# Load all sheets into a single DataFrame
sheet_names = ['Resultados_TF_40Hz', 'Resultados_TF_13Hz', 'Resultados_F_40Hz', 'Resultados_F_13Hz', 'Resultados_T_40Hz', 'Resultados_T_13Hz']
dataframes = []

for sheet in sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet)
    # Extract domain and speed from the sheet name
    if 'TF' in sheet:
        domain = 'TF'
    elif 'F' in sheet:
        domain = 'F'
    else:
        domain = 'T'
    speed = sheet.split('_')[-1]
    # Add domain and speed columns to the dataframe
    df['Domain'] = domain
    df['Speed'] = speed
    dataframes.append(df)

# Concatenate all dataframes
df_all = pd.concat(dataframes, ignore_index=True)

# Ensuring column names exist and type conversion
df_all['Feature'] = df_all['Feature'].astype(str)
df_all['Acc'] = df_all['Acc'].astype(float)
df_all['Método'] = df_all['Método'].astype(str)
df_all['Channel'] = df_all['Channel'].astype(str)
df_all['Domain'] = df_all['Domain'].astype(str)

# Grouping and calculating mean accuracy by Feature, Channel, and Domain
mean_acc_by_architecture = df_all.groupby(['Feature', 'Channel', 'Domain'])['Acc'].mean().reset_index()

# Define the desired order of domains
domain_order = ['T', 'F', 'TF']

# Map the Domain to the specified order
mean_acc_by_architecture['Domain'] = pd.Categorical(mean_acc_by_architecture['Domain'], categories=domain_order, ordered=True)

# Order the features by their domain
ordered_features = mean_acc_by_architecture.sort_values('Domain')['Feature'].unique()
mean_acc_by_architecture['Feature'] = pd.Categorical(mean_acc_by_architecture['Feature'], categories=ordered_features, ordered=True)

# Pivot the data for a heatmap
heatmap_data = mean_acc_by_architecture.pivot('Feature', 'Channel', 'Acc')

# Plotting the heatmap
fig, ax = plt.subplots(figsize=(14, 10))
heatmap = sns.heatmap(heatmap_data, cmap="YlGnBu", annot=True, fmt=".2f", linewidths=.5, ax=ax)

# Define font sizes
title_fontsize = 20
label_fontsize = 14
tick_label_fontsize = 12  # Increased tick label font size
legend_fontsize = 14

# Add and customize colorbar
colorbar = heatmap.collections[0].colorbar
colorbar.set_label('Mean Accuracy')
colorbar.set_ticks([0, 0.5, 1])  # Customize tick values according to your data range
colorbar.ax.tick_params(labelsize=legend_fontsize)  # Customize colorbar tick labels size

plt.title('Mean Accuracy by Channel and Feature', fontsize=title_fontsize)
plt.xlabel('Channel', fontsize=label_fontsize)
plt.ylabel('Feature', fontsize=label_fontsize)

# Increase tick label font size
ax.tick_params(axis='x', labelsize=tick_label_fontsize)
ax.tick_params(axis='y', labelsize=tick_label_fontsize)

# Color dictionary for the domains
domain_colors = {'T': '#8c564b', 'F': '#e377c2', 'TF': '#7f7f7f'}

# Get the y-axis labels and color them according to the domain
for label in ax.get_yticklabels():
    feature = label.get_text()
    domain = df_all[df_all['Feature'] == feature]['Domain'].iloc[0]
    label.set_color(domain_colors[domain])

# Adding custom legend for domain colors
legend_patches = [mpatches.Patch(color=color, label=domain) for domain, color in domain_colors.items()]
plt.legend(handles=legend_patches, title='Domain', bbox_to_anchor=(1.05, 1), loc='lower right')

plt.tight_layout()
plt.show()

file_name = './imagens_artigo/Mean_Accuracy_by_Channel_and_Feature_heatmap_ordered_with_colorbar_and_legend.png'
fig.savefig(file_name, dpi=600)
plt.close()
```
